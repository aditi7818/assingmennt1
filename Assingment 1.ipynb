```python
Question: explain with an example each when to use a for loop and a while loop.
Answer: for loop: A for loop is a control flow statement that executes code repeatedly for a particular number of iterations. in this control flow statement,
       the keyword used is for. the for loop is used when the number of iterations is already known.
    example of for loop:
        n= [1,2,3]
        for i in n:
            print(i)
while loop: a loop that executes a single statement or a group of statements for the given true condition. the keyword used to repersent this loop is "while"
            a "while" loop is used when the number of iterations is unknown. the statement repeats itself till the boolean value becomesfalse. in a while loop, 
           the condition is tested at the start, also known as the pre-tested loop.
        example of while loop:
            i=1 
            while i<6:
                print(i)
                i=i+1
```

```python
Question: Write a program to filter count vowels in the below-given string.
string="i want to become a data scientist"
Answer:
a=list("i want to become a data scientist")
def vowel(x):
    v=('a','e','i','o','u')
    return x in v
b=list(filter(vowel,a))
print("vowels:",b)
print("no of vowels:",len(b))
```
